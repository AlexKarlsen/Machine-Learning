```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import colorcet as cc
from load_data import loadVectors
from sklearn.decomposition import PCA, KernelPCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
x_train, y_train, x_validation, y_validation, x_test = loadVectors()
```

```python
pca = PCA(n_components=2)
PC = pca.fit_transform(x_train)
```

```python
x = []
y = []
for i in PC:
    x.append(i[0])
    y.append(i[1])
    
plot_aspect = 1.2
plot_height = 6.0
plot_width = int(plot_height*plot_aspect)
cbar_max = 39
cbar_min = 0
cbar_step = 1.0
    
size = []
for i in PC:
    new_size = i/len(PC)
    size.append(10000*new_size+100)
fig, ax = plt.subplots(figsize=(plot_height,plot_width),dpi= 120, facecolor='w', edgecolor='k')
im = ax.scatter(x, y, c=y_train, alpha = 0.5, cmap = cc.m_rainbow_bgyr_35_85_c72_r)
fig.colorbar(im, ax=ax, ticks=np.arange(cbar_min, cbar_max+cbar_step, cbar_step))
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA')
plt.show()
```

```python
kpca = KernelPCA(n_components=2, kernel='rbf')
KPC = kpca.fit_transform(x_train, y_train)
```

```python
x = []
y = []
for i in KPC:
    x.append(i[0])
    y.append(i[1])
    
plot_aspect = 1.2
plot_height = 6.0
plot_width = int(plot_height*plot_aspect)
cbar_max = 39
cbar_min = 0
cbar_step = 1.0
    
size = []
for i in PC:
    new_size = i/len(PC)
    size.append(10000*new_size+100)
fig, ax = plt.subplots(figsize=(plot_height,plot_width),dpi= 120, facecolor='w', edgecolor='k')
im = ax.scatter(x, y, c=y_train, alpha = 0.5, cmap = cc.m_rainbow_bgyr_35_85_c72_r)
fig.colorbar(im, ax=ax, ticks=np.arange(cbar_min, cbar_max+cbar_step, cbar_step))
plt.xlabel('KPC1')
plt.ylabel('KPC2')
plt.title('KPCA')
plt.show()
```
