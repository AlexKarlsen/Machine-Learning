# Kaggle AlexNet

```python
import numpy as np
from datetime import datetime
from scipy import ndimage
from PIL import Image
import matplotlib.pyplot as plt
import tensorflow as tf
import pandas as pd
```

```python
# Prepare data
img_height, img_width = 256, 256

path_train_images = './Data/Train/TrainImages/'
path_validation_images = './Data/Validation/ValidationImages/'
path_test_images = './Data/Test/TestImages/'

x_train = [] # Empty list
for i in range(1,5831):
    img = Image.open(path_train_images + 'Image' + str(i) + '.jpg') # Loading Image<x>.jpg
    x_train.append(np.array(img)/ 255.0)
x_train = np.asarray(x_train)
y_train = pd.read_csv('./Data/Train/trainLbls.csv',header=None)
y_train = np.ravel(y_train)

x_val = [] # Empty list
for i in range(1,2299):
    img = Image.open(path_validation_images + 'Image' + str(i) + '.jpg') # Loading Image<x>.jpg
    x_val.append(np.array(img)/ 255.0)
x_val = np.asarray(x_val)
y_val = pd.read_csv('./Data/Validation/valLbls.csv',header=None)
y_val = np.ravel(y_val)
```

```python
y_train.shape
```

```python
x_val.shape
```

```python
y_val.shape
```

```python
# Define model
model = tf.keras.Sequential(
    [ tf.keras.layers.Conv2D(filters=16, kernel_size=(5, 5), activation=tf.nn.relu, input_shape=(img_height, img_width, 3))
    , tf.keras.layers.AveragePooling2D()
    , tf.keras.layers.Conv2D(filters=16, kernel_size=(3, 3), activation=tf.nn.relu)
    , tf.keras.layers.AveragePooling2D()
     , tf.keras.layers.Conv2D(filters=16, kernel_size=(3, 3), activation=tf.nn.relu)
    , tf.keras.layers.AveragePooling2D()
    , tf.keras.layers.Flatten()
    , tf.keras.layers.Dense(units=120, activation=tf.nn.relu)
    , tf.keras.layers.Dense(units=84, activation=tf.nn.relu)
    , tf.keras.layers.Dense(units=30, activation=tf.nn.softmax)
    ]
)

model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
tensorboard = tf.keras.callbacks.TensorBoard(log_dir=f'logs/{datetime.now():%Y%m%d%H%M%S}')
# Open TensorBoard using $'tensorboard --logdir=logs/'

model.fit(x_train, y_train, batch_size = 64, epochs=10, callbacks=[tensorboard])
```

```python
model.evaluate(x_val, y_val)
```

```python
x_test = [] # Empty list
for i in range(1,3461):
    img = Image.open(path_test_images + 'Image' + str(i) + '.jpg') # Loading Image<x>.jpg
    x_test.append(np.array(img)/ 255.0)
x_test = np.asarray(x_test)
```

```python
x_test.shape
```

```python
predictions = model.predict_classes(x_test)
```

```python
predictions = pd.Series(predictions, name="Label",dtype='int32')
submission = pd.concat([pd.Series(range(1 ,3461) ,name = "ID"),   predictions],axis = 1)
submission.to_csv("kaggle_results.csv",index=False)
submission.head(10)
```
